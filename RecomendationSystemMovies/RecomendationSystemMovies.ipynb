```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("/content/movies.csv",usecols=["movieId","title"])
movies.head()
```

```python
rating=pd.read_csv("/content/ratings.csv",usecols=["userId","movieId","rating"])
rating.head()
```

```python
rating.shape
```

```python
reating_users=rating.pivot(index="movieId",columns="userId",values="rating").fillna(0)
reating_users.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
mat_movies=csr_matrix(reating_users.values)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric="cosine",algorithm="brute",n_neighbors=20)
model.fit(mat_movies)
```

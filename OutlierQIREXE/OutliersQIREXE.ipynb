```python
import pandas as pd
```

```python
data=pd.read_csv("/content/weight-height.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(data.Height,bins=20,rwidth=0.8)
plt.xlabel("Heigt")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(data.Weight,bins=20,rwidth=0.8)
plt.xlabel("Weight")
plt.ylabel("count")
plt.show()
```

```python
q1=data.Height.quantile(.25)
q3=data.Height.quantile(.75)
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit=q3+IQR*1.5
lower_limit=q1-IQR*1.5
upper_limit,lower_limit
```

```python
data[(data.Height>upper_limit)|(data.Height<lower_limit)]
```

```python
new_data_hight=data[(data.Height<upper_limit)|(data.Height>lower_limit)]
new_data_hight.sample(5)
```

```python
new_data_hight.shape
```

```python
Q1 = data.Weight.quantile(0.25)
Q3 = data.Weight.quantile(0.75)
Q1, Q3
```

```python
IQR2=Q3-Q1
IQR2
```

```python
upper_limit2=Q3+1.5*IQR2
lower_limit2=Q1-1.5*IQR2
upper_limit2,lower_limit2
```

```python
data[(data.Weight<lower_limit2)|(data.Weight>upper_limit2)]
```

```python
new_data_wight=data[(data.Weight<upper_limit2)&(data.Weight>lower_limit2)]
new_data_wight
```

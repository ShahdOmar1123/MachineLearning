```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('/content/bhp (1).csv')
data.head()
```

```python
data.price_per_sqft.describe()
```

```python
upper_limit=data.price_per_sqft.quantile(0.999)
lower_limit=data.price_per_sqft.quantile(0.001)
upper_limit,lower_limit
```

```python
data[(data.price_per_sqft>upper_limit)|(data.price_per_sqft<lower_limit)]
```

```python
new_data=data[(data.price_per_sqft<upper_limit)&(data.price_per_sqft>lower_limit)]
new_data.head()
```

```python
new_data.describe()
```

```python
data.shape
```

```python
new_data.shape
```

```python
max_limit = new_data.price_per_sqft.mean() + 4*new_data.price_per_sqft.std()
min_limit = new_data.price_per_sqft.mean() - 4*new_data.price_per_sqft.std()
max_limit, min_limit
```

```python
new_data[(new_data.price_per_sqft>max_limit)|(new_data.price_per_sqft<lower_limit)]
```

```python
data3=new_data[(new_data.price_per_sqft<max_limit)&(new_data.price_per_sqft>min_limit)]
data3.shape
```

```python
plt.hist(data3.price_per_sqft,bins=20,rwidth=0.8)
plt.xlabel("Price per square ft")
plt.ylabel("Count")
plt.show()
```

```python
from scipy.stats import norm
import numpy as np

plt.hist(data3.price_per_sqft, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(-5000, data3.price_per_sqft.max(), 100)
plt.plot(rng, norm.pdf(rng,data3.price_per_sqft.mean(),data3.price_per_sqft.std()))

```

```python
new_data['Z_Score'] = (
    new_data['price_per_sqft'] - new_data['price_per_sqft'].mean()
) / new_data['price_per_sqft'].std()

new_data.sample(10)
```

```python
outliers=new_data[(new_data.Z_Score<-4)|(new_data.Z_Score>4)]
outliers.shape
```
